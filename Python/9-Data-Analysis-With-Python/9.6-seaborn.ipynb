#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

```python
#Basic plotting with seaborn

tips=sns.load_dataset('tips')
```

```python
tips
```

Scatter Plot

```python
## create a scatterplot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Total bill vs tips")
plt.show()
```

Line Plot

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill vs size")
plt.show()
```

Categorical Plots

1-Bar Plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Total bill for each day")
```

2-Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

3-Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

Histograms

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

KDE PLOT- KERNEL DENSITY ESTIMATOR

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

Pair Plot

```python
sns.pairplot(tips)
```

Heatmap

```python
corr=tips[['total_bill','tip','size']].corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
