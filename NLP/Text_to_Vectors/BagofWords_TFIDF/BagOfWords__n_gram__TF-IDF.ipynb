### Bag Of Words Implementation

```python
import pandas as pd
message=pd.read_csv("SMSSpamCollection.txt",sep="\t",names=["label","message"])
```

```python
message
```

```python
## Data Cleaning and preprocessing
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize,wordpunct_tokenize
from nltk.stem import WordNetLemmatizer, SnowballStemmer,PorterStemmer
import re
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]

for i in range(0,len(message)):
    review=re.sub('[^a-zA-z]',' ',message['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer=CountVectorizer(lowercase=True,stop_words='english',max_features=100,binary=True)
```

```python
X=count_vectorizer.fit_transform(corpus).toarray()
```

```python
count_vectorizer.vocabulary_
```

```python
X
```

### n_grams Implementation

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer=CountVectorizer(lowercase=True,stop_words='english',max_features=2500,binary=True,ngram_range=(1,3))
```

```python
X=count_vectorizer.fit_transform(corpus).toarray()
```

```python
count_vectorizer.vocabulary_
```

### TF-IDF Implementation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=100)
```
