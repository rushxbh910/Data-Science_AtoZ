### Support vector classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue=y)
```

```python
#train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25,random_state=10 )
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear') #since points are linearly separable we use linear svc/svm
```

```python
svc.fit(X_train,y_train)
```

```python
y_hat_test=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
```

### RBF kernel

```python
# using rbf kernel for non linearly seperable dataset
```

```python
X,y=make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue=y)
```

```python
#train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25,random_state=10 )
```

```python
rbf=SVC(kernel='rbf') #since points are NOT linearly separable we use rbf svc/svm
```

```python
rbf.fit(X_train,y_train)
```

```python
y_hat_test= rbf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
```

### Polynomial Kernel

```python
# using polynomial kernel for non linearly seperable dataset
```

```python
poly=SVC(kernel='poly') #since points are NOT linearly separable we use rbf svc/svm
poly.fit(X_train,y_train)
y_hat_test= poly.predict(X_test)

print(classification_report(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
```

### Sigmoid Kernel

```python
# using sigmoid kernel for non linearly seperable dataset
```

```python
sigmoid=SVC(kernel='sigmoid') #since points are NOT linearly separable we use rbf svc/svm
sigmoid.fit(X_train,y_train)
y_hat_test= sigmoid.predict(X_test)

print(classification_report(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
```

we can compare the results of poly and rbf and decide what to use

## Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf','sigmoid','poly']
}

grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_hat_test=grid.predict(X_test)
```

```python
print(classification_report(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
```
