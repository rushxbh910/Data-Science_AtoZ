Naive Bayes Machine Learning Implementation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
#train test split
X_Train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
```

```python
nb.fit(X_Train,y_train)
```

```python
y_hat_test=nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test,y_hat_test))
print(confusion_matrix(y_test,y_hat_test))
print(classification_report(y_test,y_hat_test))
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X=df[['total_bill','tip','sex','smoker','day','size']]
y=df[['time']]
```

```python
X
```

```python
X_Train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
y_train
```

```python
X_Train['sex']=le1.fit_transform(X_Train['sex'])
X_Train['smoker']=le2.fit_transform(X_Train['smoker'])
```

```python
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
```

```python
y_train['time']=le3.fit_transform(y_train['time'])
y_test['time']=le3.transform(y_test['time'])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
```

```python
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('one hot',OneHotEncoder(drop='first'),[4])],remainder='passthrough')

X_Train=ct.fit_transform(X_Train)
X_test=ct.transform(X_test)
```
