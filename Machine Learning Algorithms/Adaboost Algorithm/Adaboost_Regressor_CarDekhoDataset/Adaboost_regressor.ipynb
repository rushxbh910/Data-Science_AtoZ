Used Car Price Prediction
1) Problem statement.
- This dataset comprises used cars sold on cardehko.com in India as well as important features of these cars.
If user can predict the price of the car based on input features.
Prediction results can be used to give new seller the price suggestion based on market condition.
2) Data Collection.
- The Dataset is collected from scrapping from cardheko webiste
- The data consists of 13 column and 15411 rows.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(columns=['car_name','brand'],inplace=True)
```

```python
num_features=[feature for feature in df.columns if df[feature].dtype !='O']
print("no. of numerical features: ",len(num_features))

cat_features=[feature for feature in df.columns if df[feature].dtype =='O']
print("no. of categorical features: ",len(cat_features))

discrete_features=[feature for feature in num_features if len(df[feature].unique())<25]
print("no. of discrete features: ",len(discrete_features))

continuos_features=[feature for feature in num_features if feature not in discrete_features]
print("no. of categorical features: ",len(cat_features))
```

```python
df.head()
```

```python
#independant and dependant features

X=df.iloc[:,:-1]
y=df[['selling_price']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()

df['model']=le1.fit_transform(df[['model']])
```

```python
df.info()
```

```python
# feature scaling and encoding
#creating column transfer
cat_features=X.select_dtypes(include='object').columns
numeric_features=X.select_dtypes(exclude='object').columns
```

```python
ohe=OneHotEncoder(drop='first')
scaler=StandardScaler()

preprocessor=ColumnTransformer(
    [
        ("onehotencoder",ohe,cat_features),
        ("scaler",scaler,numeric_features)
    ],remainder='passthrough'
)
```

```python
X=preprocessor.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

### Model selection, Hyperparameter tuning and Model selection

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python
models={
    "Linear Reg ":LinearRegression(),
    "K Nearest neighbours":KNeighborsRegressor(),
    "Decision tree": DecisionTreeRegressor(),
    "Random forest":RandomForestRegressor(),
    "Adaboost Regressor": AdaBoostRegressor()
}
```

```python
for model in range(len(list(models))):
    m=list(models.values())[model]
    m.fit(X_train,y_train)

    #predict test data
    y_hat_test=m.predict(X_test)

    #performance check
    print("_"*100)
    print(f"r2 score of {list(models.keys())[model]}",r2_score(y_test,y_hat_test))
    print(f"mse of {list(models.keys())[model]}",mean_squared_error(y_test,y_hat_test))
    print(f"mae of {list(models.keys())[model]}",mean_absolute_error(y_test,y_hat_test))
    print(f"rmse {list(models.keys())[model]}",root_mean_squared_error(y_test,y_hat_test))
    print("_"*100)
```

```python
#hyper parameter tuning knn and random forest

knn_params={"n_neighbors":[2,3,10,20,40,50]}
rf_params={
    "max_depth":[5,8,15,None,10],
    "max_features":[5,7,"auto",8],
    "min_samples_split":[2,8,15,20],
    "n_estimators":[100,200]
}
ada_params={
    "n_estimators":[50,60,70,80],
}
```

```python
gridcv_models=[
    ('KNN',KNeighborsRegressor(),knn_params),
    ('rf',RandomForestRegressor(),rf_params),
    ('adaboost_reg',AdaBoostRegressor(),ada_params)
]
```

```python
from sklearn.model_selection import GridSearchCV

model_params={}
for name,model,params in gridcv_models:
    grid=GridSearchCV(estimator=model,param_grid=params,n_jobs=-1,cv=5)
    grid.fit(X_train,y_train)
    model_params[name]=grid.best_params_

for model_name in model_params:
    print(f"best parameters for {model_name}")
    print(model_params[model_name])
```

```python
models={
    "K Nearest neighbours":KNeighborsRegressor(n_neighbors=3),
    "Random forest":RandomForestRegressor(max_depth= 15, max_features=5, min_samples_split= 2, n_estimators= 100),
    "Adaboost Reg":AdaBoostRegressor(n_estimators=50)
}

for model in range(len(list(models))):
    m=list(models.values())[model]
    m.fit(X_train,y_train)

    #predict test data
    y_hat_test=m.predict(X_test)

    print("_"*100)
    print(list(models.keys())[model])
    #performance check
    
    print(f"r2 score of {list(models.keys())[model]}",r2_score(y_test,y_hat_test))
    print(f"mse of {list(models.keys())[model]}",mean_squared_error(y_test,y_hat_test))
    print(f"mae of {list(models.keys())[model]}",mean_absolute_error(y_test,y_hat_test))
    print(f"rmse {list(models.keys())[model]}",root_mean_squared_error(y_test,y_hat_test))
    print("_"*100)
```
