```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating simple df
df=pd.DataFrame({'x':[0,4,9,16,25,36,49,64,81,100],
                'y':[0,2,3,4,5,6,7,8,9,10]})
```

```python
X=df[['x']]
Y=df['y']
```

```python
#Creating random values for x and y
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2 +1.5*x+2+np.random.randn(100,1)
```

```python
plt.scatter(x,y)
```

```python
#train test split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#implementing simple linear regression to compare the error and best fit line with polynomial regression

from sklearn.linear_model import LinearRegression

lin_regression=LinearRegression()
```

```python
lin_regression.fit(x_train,y_train)
```

```python
#predict x_test and get y_hat value
y_hat=lin_regression.predict(x_test)
```

```python
#check performance of the model using r2 performance metrics
from sklearn.metrics import r2_score

score=r2_score(y_test,y_hat)
print(score)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,lin_regression.predict(x_train),color='r')
```

```python
#polynomial reg model
```

```python
## applying polynomial transformation
##(to keep increasing the degree and checking the fit)

from sklearn.preprocessing import PolynomialFeatures #polynomial features is used for transformation/to increase the degree
```

```python
poly=PolynomialFeatures(degree=2, include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly #output=bias+beta1*x1+beta2*x2^2
```

```python
x_test_poly
```

```python
#train poly reg model (similar to linear reg)

pol_reg=LinearRegression()
pol_reg.fit(x_train_poly,y_train)
y_hat_poly=pol_reg.predict(x_test_poly)
```

```python
from sklearn.metrics import r2_score
score_poly=r2_score(y_test,y_hat_poly)
print(score_poly)
```

```python
print(pol_reg.coef_)
```

```python
print(pol_reg.intercept_)
```

```python
plt.scatter(x_train,pol_reg.predict(x_train_poly))
plt.scatter(x_train,y_train)
```

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)
    
    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("poly_features",poly_features),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(x_train,y_train) ## ploynomial and fit of linear reression
    y_pred_new=poly_regression.predict(X_new)
    #plotting prediction line
    plt.plot(X_new, y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
    plt.plot(x_train, y_train, "b.", linewidth=3)
    plt.plot(x_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
```

```python
poly_regression(5)
```
